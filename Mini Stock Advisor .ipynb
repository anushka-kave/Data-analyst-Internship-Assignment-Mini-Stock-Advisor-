```python
# Mini Stock Advisor 

import pandas as pd

# Create 10-day dataset
data = pd.DataFrame({
    "Day": range(1, 11),
    "Open": [100,103,105,104,101,102,104,100,98,99],
    "Close": [103,105,104,101,102,104,100,98,99,102]
})

# Compute percentage change
data["percentage_change"] = ((data["Close"] - data["Open"]) / data["Open"]) * 100

# Determine positive/negative
data["positive"] = data["percentage_change"] > 0

# Apply rule-based system
signals = []

for i in range(len(data)):
    if i == 0:
        signals.append("HOLD")  # Day 1 has no yesterday
        continue
    
    today = data.loc[i, "percentage_change"]
    yesterday = data.loc[i-1, "percentage_change"]
    today_pos = data.loc[i, "positive"]
    yesterday_pos = data.loc[i-1, "positive"]

    # BUY Rules
    if (today > 2 and yesterday_pos) or (today_pos and yesterday_pos):
        signals.append("BUY")
    
    # SELL Rules
    elif (today < -2) or (not today_pos and not yesterday_pos):
        signals.append("SELL")
    
    # HOLD
    elif -2 <= today <= 2:
        signals.append("HOLD")
    
    else:
        signals.append("HOLD")

data["Signal"] = signals

# Print final table
print(data)
```
